### Handling PDF Issues/Challenges

Purpose of this PDFs are not notoriously difficult ot parse because they:
- Store text in complex ways (not just simple text)
- can have formatting issues
- May contain scanned images (requiring OCR)
- Often have extracted artifacts 

```python
raw_pdf_text = """Company finacial report



    The finacial performance ﬁscal year 2024
    shows significant growth in profitability


Revenue Increse by 25%

    The company's efficiency improved due to workflow 
optimization


page 1 of 10

"""


# Apply cleaning functions

def clean_text(text):
    # Remove excessive whitespace
    text = " ".join(text.split())

    #Fix ligatures
    text = text.replace("ﬁ", "fi")
    return text


cleaned = clean_text(raw_pdf_text)
print("BEFORE : ")
print(repr(raw_pdf_text))
print("AFTER : ")
print(repr(cleaned))
```

```python
from typing import List
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
```

```python
# Class

class SmartPDFProcessor:
    """Advance PDFs Processig with error handling"""

    def __init__(self, chunk_size=1000, chunk_overlap=100):
        self.chunk_size = chunk_size,
        self.chunk_overlap = chunk_overlap,
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size = chunk_size,
            chunk_overlap = chunk_overlap,
            separators = [" "]
        )


    def process_pdf(self, pdf_path: str)->List[Document]:
        """Process PDF with smart chunking and metadata enchancement"""

        # Load the PDF
        loader = PyPDFLoader(pdf_path)
        pages = loader.load()

        # Process each page
        processed_chunks = []

        for page_num, page in enumerate(pages):
            ## Clean the text
            cleanned_text = self._clean_text(page.page_content)

            # Skip nearly empty pages
            if len(cleanned_text.strip()) < 50:
                continue
        
            # Create chunks with enchanced metadata
            chunks = self.text_splitter.create_documents(
                texts = [cleanned_text],
                metadatas = [{
                    **page.metadata,
                    "page": page_num+1,
                    "total_pages": len(pages),
                    "chunk_method": "Smart Chunk Method",
                    "char_count": len(cleanned_text)
                }]
            )

            processed_chunks.extend(chunks)

        return processed_chunks

    
    def _clean_text(self, text:str) -> str:
        """Clean Extracted text"""

        # Remove Excessive whitespace
        text = " ".join(text.split())

        return text
```

```python
preprocessor = SmartPDFProcessor()
```

```python
preprocessor
```

```python
## Process the PDF
try:
    smart_chunks = preprocessor.process_pdf('data/pdf/Introduction to Agents.pdf')
    print(f"Processed into {len(smart_chunks)} chunks")

    # Show enchanced metadata
    if smart_chunks:
        print('\nSample Chunk Metadata :')
        for key, value in smart_chunks[0].metadata.items():
            print(f" {key} : {value}")
        
        print("\nSample Chunk Content :")
        print(f" {smart_chunks[0].page_content[:150]}...")

except Exception as e:
    print(f"Error : {e}")
```
